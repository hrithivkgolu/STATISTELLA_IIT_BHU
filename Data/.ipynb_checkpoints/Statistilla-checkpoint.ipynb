```python
import pandas as pd
games=pd.read_csv('/games.csv')
games_details=pd.read_csv('/games_details.csv')
players=pd.read_csv('/players.csv')
ranking=pd.read_csv('/ranking.csv')
teams=pd.read_csv('/teams.csv')
```

```python
ranking['STANDINGSDATE']=pd.to_datetime(ranking['STANDINGSDATE'],errors='coerce')
```

```python
games['TOTAL_POINTS'] = games['PTS_home'] + games['PTS_away']
```

```python
season_trend = (
    games
    .groupby('SEASON')['TOTAL_POINTS']
    .mean()
    .reset_index()
)
```

```python
team_performance = (
    ranking
    .groupby('TEAM_ID')[['W', 'L']]
    .sum()
    .reset_index()
)

team_performance['WIN_PCT'] = (
    team_performance['W'] /
    (team_performance['W'] + team_performance['L'])
)
```

```python
team_performance = team_performance.merge(
    teams[['TEAM_ID', 'NICKNAME']],
    on='TEAM_ID',
    how='left'
)
```

```python
player_stats = (
    games_details
    .groupby('PLAYER_ID')[['PTS', 'AST', 'REB']]
    .mean()
    .reset_index()
)
```

```python
player_stats = player_stats.merge(
    players[['PLAYER_ID', 'PLAYER_NAME']],
    on='PLAYER_ID',
    how='left'
)
```

```python
games_details['PTS'] = pd.to_numeric(games_details['PTS'], errors='coerce').fillna(0)
games_details['MIN'] = pd.to_numeric(games_details['MIN'], errors='coerce').fillna(0)

player_minutes = (
    games_details
    .groupby('PLAYER_ID')[['PTS', 'MIN']]
    .mean()
    .reset_index()
)

player_minutes['PTS_PER_MIN'] = (
    player_minutes['PTS'] / player_minutes['MIN']
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(season_trend['SEASON'], season_trend['TOTAL_POINTS'])
plt.title("Season-wise Average Points")
plt.xlabel("Season")
plt.ylabel("Avg Points")
plt.show()
```

```python
top_teams = team_performance.sort_values(
    'WIN_PCT', ascending=False
).head(10)

plt.figure(figsize=(8,5))
plt.barh(top_teams['NICKNAME'], top_teams['WIN_PCT'])
plt.title("Top 10 Teams by Win Percentage")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(player_minutes['MIN'], player_minutes['PTS'])
plt.xlabel("Avg Minutes")
plt.ylabel("Avg Points")
plt.title("Player Efficiency Distribution")
plt.show()
```

```python
import plotly.express as px

fig = px.line(
    season_trend,
    x='SEASON',
    y='TOTAL_POINTS',
    title="Season-wise Scoring Trend"
)
fig.show()
```
