```python
#team wise player data
import pandas as pd
```

```python
player = pd.read_csv("players.csv")
team = pd.read_csv("teams.csv")
merged_df = pd.merge(team, player, on='TEAM_ID', how='left')
merged_df = merged_df.drop_duplicates(subset=['PLAYER_ID'], keep='first')
merged_df.to_csv('final_teams.csv', index=False)
```
