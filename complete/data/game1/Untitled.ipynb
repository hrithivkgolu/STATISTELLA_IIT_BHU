```python
import pandas as pd
import matplotlib.pyplot as plt
meanl = []

for i in range(2003,2022):
    df = pd.read_csv(f"g{i}.csv")
    df["total_points"] = df["PTS_home"]+df["PTS_away"]
    h = df["total_points"].max()
    l = df["total_points"].min()
    m = df["total_points"].mean()
    a = [i,float(round(m,2)),float(round(h,2)),float(round(l,2))]
    meanl.append(a)

file = pd.DataFrame(meanl,columns = ["Year","Mean","Highest","Lowest"])
file.to_csv("MHL.csv")
# season_score = 
# print(min(low))
# plt.plot(years, values, label="mean", color='blue', linestyle='-')
# plt.plot(years, high, label='highest', color='red', linestyle='--')
# plt.plot(years, low, label='lowest', color='green', linestyle='-.')

# # 3. Add customizations
# plt.title("Multi-Line Plot Example")
# plt.xlabel("X-axis Value")
# plt.ylabel("Y-axis Value")
# plt.legend(loc='upper right')
# plt.grid(True)

# plt.show()
```
